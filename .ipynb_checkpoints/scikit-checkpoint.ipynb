```python
# Importando importantes pacotes para importação dados que importam.
import pandas as pd
import numpy as np
```

### Dados

```python
# importando dados
label_test = pd.read_csv('gender_submission.csv') # labels corretos do test set
train = pd.read_csv('train.csv') 
test = pd.read_csv('test.csv') 
# label_test.head()  
# train.head()
test.head()
```

Data dictionary



## Pré-processamento

```python
train.head()
```

```python
# verificando valores NaN (not a number)
print(train.info()) #Age, Cabin e Embarked possuem valores nulos
print()
print(test.info())
```

```python
# Coluna cabin ->  muitos valores nulos, vamos retirar a coluna
train.drop('Cabin', inplace=True, axis=1) # inplace pra alterar direto sem ter que armazenar, axis=1 pra coluna
test.drop('Cabin', inplace=True, axis=1) # muitos valores Nan, dropamos a coluna
```

```python
# Coluna Age
print(train['Age'].isnull().sum())
mean_age = train['Age'].mean()
print(mean_age)
train['Age'].fillna(mean_age, inplace = True) # substituindo Nan pela média
print(train['Age'].isnull().sum())
mean_age_test = test['Age'].mean()
test.fillna(mean_age_test, inplace = True) # substituindo Nan pela média
```

```python
# Coluna Embarked
train['Embarked'].value_counts() # S tem maior frequencia
train['Embarked'].isnull().sum() # temos apenas dois valores faltantes
train['Embarked'].fillna('S', inplace = True) # substituimos Nan pela mediana
test['Embarked'].isnull().sum() # sem nan no test set
```

```python
# Colunas não necessárias
train.drop('Name', inplace=True, axis=1)
test.drop('Name', inplace=True, axis=1)
train.drop('Ticket', inplace=True, axis=1)
test.drop('Ticket', inplace=True, axis=1)
```

```python
train.head()
```

```python
# Variáveis categóricas para numéricas:
from sklearn.preprocessing import LabelEncoder
labelencoder_train = LabelEncoder()
train[:,4] = labelencoder_train.fit_transform(train[:,4])
```

```python
train["Sex_cat"] = train["Sex"].cat.codes
train["Embarked_cat"] = train["Embarked"].cat.codes
```

```python
# Escalonamento: 
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler() 
train = scaler.fit_transform(train)
```
